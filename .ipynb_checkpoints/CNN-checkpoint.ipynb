```python
def padSet(batch, pad):
    #returns list of 1D arrays containing padded batch images
    #batch is the input images, pad is the padding size
    new = []

    for b in range(len(batch)):
        img = batch[b]
        pads = np.zeros((np.shape(img)[0] + pad * 2, np.shape(img)[1] + pad * 2))
        pads[pad:-pad, pad:-pad] = img
        new.append(pads)
    return new
```

```python
def padImg(img, pad):
    #pads single image for convolution
    pads = np.zeros((np.shape(img)[0] + pad * 2, np.shape(img)[0] + pad * 2))
    pads[pad:-pad, pad:-pad] = img
    return pads
```

```python
def relu(x):
    x[x < 0] = x[x<0]*0.01
    return x
x = np.array([-1, -2, 4, 3, -.12, 0])
relu(x)
```

```python
def makefilter(size, scale,zeros=False):
    if zeros:
        return np.zeros(size)
    if not zeros:
        stddev = scale/np.sqrt(np.prod(size))
        return np.random.normal(loc = 0, scale = stddev, size = size)
makefilter((3,3), 3)
```

```python
def makeWeights(fil1, fil2, dense, out, zeros=False):
    #fil1, fil2 = tuples w 1st arg: numFilters, 2nd arg: dims, 3rd arg: scale of stddev
    layers = []
    cache = []
    for i in range(fil1[0]):
        f = makefilter((fil1[1], fil1[1]), fil1[2], zeros=zeros)
        cache.append(f)
    layers.append(cache)
    
    cache = []
    for n in range(fil2[0]):
        f = makefilter((fil1[0], fil2[1], fil2[1]), fil2[2], zeros=zeros)
        cache.append(f)
    layers.append(cache)
    
    soft = makefilter((dense, out), 1,zeros=zeros)
    layers.append(soft)
    
    return layers
l = makeWeights((8, 5, 3), (16, 5, 3), (7*7*16), 10)
print(l[0][0])
print(np.shape(np.asarray(l[0])), np.shape(np.asarray(l[1])), np.shape(np.asarray(l[2])))
```

```python
def maxpool(convout, winsize, stride):
    #calculate output size
    dim = int((np.shape(convout[0])[0] - winsize) / stride) + 1
    poolout, allinds = [], []
    #for number of filtered outputs from convolution
    for f in range(len(convout)):
        img = convout[f]
        out = np.zeros((dim, dim))
        rowindx = []
        for row in range(dim):
            colindx = []
            for col in range(dim):
                wind = img[(stride*row):(stride*row + winsize), (stride*col):(stride*col + winsize)]
                
                #find the index of the max value as a result of pooling
                arg = np.argmax(wind)
                if arg == 0:
                    indx = stride*row, stride*col
                elif arg == 1:
                     indx = stride*row, stride*col + winsize - 1
                elif arg == 2:
                    indx = stride*row + winsize - 1, stride*col
                elif arg == 3:
                    indx = stride*row + winsize - 1, stride*col + winsize - 1
                #add index as a tuple to list
                #assign max value to corresponding place in output matrix
                out[row, col] = np.amax(wind)
                colindx.append(indx)
            rowindx.append(colindx)
        #add list of tuples for current output img to list
        allinds.append(rowindx)
        #add pooled img to list
        #print(np.shape(out))
        poolout.append(out)
    return poolout, allinds
```

```python
def convlayer(img, filters, pad):
    filtered = []
    for fil in range(len(filters)):
        out = conv(img, filters[fil], pad)
        filtered.append(relu(out))
    return filtered
```

```python
def convdepth(imgset, filters, padding):
    #imgset: list of 32 14x14 (dims 32x14x14), filters: list of 64 5x5 (64x5x5)
    filtered = []
    padded = padSet(imgset, padding)
    padimg = np.asarray(padded)
    for filt in filters:
        flip = np.flipud(np.fliplr(filt))
        out = np.zeros_like(imgset[0])
        fildim = np.shape(filt)[1]

        for row in range(np.shape(imgset)[1]):
            for col in range(np.shape(imgset)[2]):
                out[row, col] = np.sum(flip * padimg[:, row:row + fildim, col:col + fildim])
        filtered.append(relu(out))
    return filtered
```

```python
def dropout(flatimg, p):
    #returns the flattened dense layer and list of all dropped neurons
    drop = np.random.binomial(1, p, size=flatimg.shape) / p
    flatimg *= drop
    return flatimg, drop
```

```python
def forwardprop(img, weights, pad, winsz, stride, p):
    #forward pass of MNIST network
    layerOuts = []
    #first convolution
    filt = convdepth(np.asarray([img]), weights[0], pad)
    #filt = np.reshape(filt, (1, 32, 28, 28))
    layerOuts.append(filt)
    
    #first pool
    pool, ind = maxpool(filt, winsz, stride)
    layerOuts.append(pool)
    
    #second convolution
    filt = convdepth(pool, weights[1], pad)
    layerOuts.append(filt)
    
    #second pool
    pool, ind2 = maxpool(filt, winsz, stride)
    layerOuts.append(pool)
    
    #flatten
    flat = np.asarray(pool).flatten()
    layerOuts.append(flat)
    
    #fully connected layer
#     denseout = relu(np.matmul(weights[2], flat))
#     layerOuts.append(denseout)
    
    #dropout layer
    #drop, dropped = dropout(denseout, p)
    
    #logits layer
    s = np.matmul(weights[2].T, flat)   #drop, weights[3]
    out = softmaxLSE(s)
    layerOuts.append(out)
    
    return layerOuts, ind, ind2
layerOuts, ind, ind2 = forwardprop(img, l, 2, 2, 2, 0.4)
print(layerOuts[-1])
print(np.shape(np.asarray(ind)))
```

```python
def backpool(convouts, poolouts, indxs):
    backfil = []
    for i in range(len(convouts)):   #0-64
        img = poolouts[i]   #7x7
        imginds = indxs[i]    #7x7
        temp = np.zeros_like(convouts[0])
        #fill in values that corresponded to max pool results
        for row in range(len(imginds)):
            rows = imginds[row]  #1x7 of tuple pairs
            for col in range(len(rows)):
                coord = rows[col] #specific tuple
                temp[coord[0], coord[1]] = img[row, col]
        backfil.append(relu(temp))
    return backfil
xx = backpool(layerOuts[2], layerOuts[3], ind2)
np.shape(xx)
```

```python
def backconv(backpoolout, weights, pad):
    kerns = np.reshape(weights, (np.shape(weights)[1], np.shape(weights)[0], np.shape(weights)[2], np.shape(weights)[3]))
    #rotates weight matrix 180 degrees
    for i in range(np.shape(kerns)[0]):
        for n in range(np.shape(kerns)[1]):
            kerns[i,n,:,:] = np.fliplr(np.flipud(kerns[i,n,:,:]))
    #backprop of convolutional layer
    conv2grad = convdepth(np.asarray(backpoolout), kerns, pad)
    
    return conv2grad
```

```python
def convgrad(backpoolout, filters, convout, pad, lrate):
    #takes 64 filters, loops through the depth of the input img
    #img should be 32x1x14x14, backpoolout is 64x1x14x14, for each of the 64 layers from the outpool layer the
    #"filter must loop through 32 layers of input img, outputs a 32x5x5 vector, 64 times
    filtered = []
    padded = padSet(convout, pad)   #
    padimg = np.asarray(padded)

    for filt in filters:    #0-64
        flip = np.flipud(np.fliplr(filt))
        out = np.zeros_like(flip)   #32x5x5
        fildepth, fildim = np.shape(filt)[0], np.shape(filt)[1]

        for dep in range(fildepth):
            for row in range(fildim):
                for col in range(fildim):
                    out[dep, row, col] = np.sum(flip * padimg[dep, row:row + fildim, col:col + fildim])
        filtered.append(lrate * out)
    return filtered
oh = convgrad(xx, l[1], layerOuts[2], 2, .01)
print(np.shape(layerOuts[2]))
print(oh[0][0])
```

```python
def backprop(img, tar, weights, outs, ind, ind2, lrate, pad):
    grad = []
    outerr = outs[-1] - tar
    
    #backwards fully connected, where "current" layer weights take into account the dropout
    ndim = (1, np.shape(outerr)[0])
    mdim = (np.shape(outs[-2])[0], 1)
    denseback = lrate * np.matmul(np.reshape(outs[-2], mdim), np.reshape(outerr, ndim))     
    grad.insert(0, denseback)
    
    #from fully connected to pooling outputs
#     outJ = np.reshape(outs[-3], (784, 1))   #3136
#     deltaK = relu(np.matmul(weights[-2], outJ))
#     ndim = (np.shape(deltaK)[0], 1)
#     mdim = (1, np.shape(outJ)[0])
#     dense2pool = lrate * np.matmul(np.reshape(deltaK, ndim), np.reshape(outJ, mdim))
#     grad.insert(0, dense2pool)
    
    #backwards pooling: set all filter values to zero except for the one that corresponds to the max value
    backfil2 = backpool(outs[2], outs[3], ind2)   #64x14x14
    
    #backwards conv2, now output matrix is all zeros except for max pool values
    #takes 64x14x14 input, 64x5x5 kernel, outputs 32x14x14
    conv2out = backconv(backfil2, weights[1], pad)   #TODO: Needs learning rate
    conv2grad = convgrad(backfil2, weights[1], outs[2], pad, lrate)
    grad.insert(0, conv2grad)
        
    #backwards pooling again
    backfil1 = backpool(outs[0], outs[1], ind)   #32x28x28

    #backwards conv
    conv1weights = np.reshape(weights[0], (1,8, 5, 5))    #1x32x5x5
    conv1out = backconv(backfil1, conv1weights, pad)    #TODO needs learning rate

    conv1grad = convgrad(backfil1, conv1weights, conv1out, pad, lrate)
    grad.insert(0, conv1grad[0])
    
    return grad
np.shape(l[1])
t = backprop(img, tartr[0], l, layerOuts, ind, ind2, .01, 2)
print(np.shape(t[2]))
```

```python
def updateweights(weights, grads):
    newWeights = []
    for layer in range(len(weights)):
        newWeights.append(np.asarray(weights[layer]) - np.asarray(grads[layer]))
    return newWeights
up = updateweights(l, t)
print(np.shape(up[2]))
```

```python
def accumGrad(imggrad, batchgrad):
    #takes gradient from one image and adds to the accumulated batch gradient
    for layer in range(len(imggrad)):
        #because the first 2 layers are a list of matrices
        if layer < 2:
            for filt in range(len(imggrad[layer])):
                batchgrad[layer] += imggrad[layer][filt]
        #because second 2 layers are just matrices
        if layer >= 2:
            batchgrad[layer] += imggrad[layer]
    return batchgrad
```

```python
def train(batch, targs, weights, lrate, prob, pad=2, winsz=2, stride=2):
    #change padding, window size and stride at YOUR OWN RISK.
    errors, grads = [], []
    numCorrect = 0
    #initialize the gradient
   
    grads = makeWeights((8, 5, 3), (16, 5, 3), (7*7*16), 10, True)  #(32, 5, 3), (64, 5, 3), (1024, (7*7*64)), 10, True)
    #Loop thru all input vectors, train
    for p in range(len(batch)):
        #test for NaNs
        for i in range(len(weights)):
            if i < 2:
                for n in range(len(weights[i])):
                    if np.any(np.isnan(weights[i][n])):
                            fig, ax = plt.subplots()
                            ax.plot(errors, marker= '.')
                            ax.set(xlabel = 'Iteration', ylabel= 'Cross Entropy Error', title= 'Training Set Error')
                            plt.show()
                            return weights, np.isnan(weights[i][n])
            if i >= 2:        
                if np.any(np.isnan(weights[i])):
                    fig, ax = plt.subplots()
                    ax.plot(errors, marker= '.')
                    ax.set(xlabel = 'Iteration', ylabel= 'Cross Entropy Error', title= 'Training Set Error')
                    plt.show()
                    return weights, np.isnan(weights[i])
        
        #reshape img to a square
        im = batch[p]
        img = im.reshape((28, 28))
        #forward pass
        outs, ind, ind2 = forwardprop(img, weights, pad, winsz, stride, prob)
        
        #backward pass, calculates gradients
        imggrad = backprop(img, targs[p], weights, outs, ind, ind2, lrate, pad)
        
        #update cumulative gradient
        grads = accumGrad(imggrad, grads)
        #print(grads[-1])
        
        #calculate network error
        error = ff.crossEntError(outs[-1], targs[p])
        errors.append(error)

        #did network guess right?
        if np.argmax(outs[-1]) == np.argmax(targs[p]):
            #print("diff:", np.amax(targs[p]) - np.amax(outs[-1]))
            numCorrect += 1
            
    #update weights
    #IF RUNNING CELL BELOW: indent weights to be w/in for loop
    #IF TRAINING FULL SET: calculate weights after batch loop completes
        weights = updateweights(weights, grads)
    
    #calculate acc and loss
    acc = (numCorrect / len(batch)) * 100
    loss =np.average(np.asarray(errors))
    
    #plot the error
    #IF TRAINING FULL SET: comment out plotting and print statements
    fig, ax = plt.subplots()
    ax.plot(errors, marker= '.')
    ax.set(xlabel = 'Iteration', ylabel= 'Cross Entropy Error', title= 'Test Set Error')
    plt.show()
    
    print("difference: ", (outs[-1] - targs[-1]))
    print("accuracy: ", acc, "%")
    print("number of correct guesses: ", numCorrect)
    
    return weights, loss, acc
wei, lo, ac = train(imgtr, tartr, l, .00001, 0.4)
```

```python
def loadweights(filename='test-cnn-mnist.pckl'):
    try:
        we = pickle.load(open(filename, 'rb'))
    except (OSError, IOError) as e:
        we = makeWeights((8, 5, 1), (16, 5, 1), (7*7*16), 10)
        pickle.dump(we, open(filename, 'wb'))
    return we
```

```python
def testNet(testimgs, testtargs, weights, lrate, prob, good=True, showNums=False, pad=2, winsz=2, stride=2):
    #uses current weights to evaluate the test set, returns average loss and average accuracy of whole test
    errors, grads, goodOnes, badOnes = [], [], [], []
    numCorrect = 0
    
    #loop through images in the test set
    for t in range(len(testimgs)):
        #reshape 1D vec to square
        im = testimgs[t]
        img = im.reshape((28, 28))
        #forward pass
        outs, ind, ind2 = forwardprop(img, weights, pad, winsz, stride, prob)
        
        #calculate network error
        error = ff.crossEntError(outs[-1], testtargs[t])
        errors.append(error)
        
        #did network guess right?
        if np.argmax(outs[-1]) == np.argmax(testtargs[t]):
            numCorrect += 1
            goodOnes.append((t, np.argmax(outs[-1])))
        if int(error) >= 1:
            badOnes.append((t, np.argmax(outs[-1])))
    #calculate accuracy
    acc = (numCorrect / len(testimgs)) * 100
    
    #visualize guesses if showNums = True
    if showNums:
        if good:
            ff.visualize(testimgs, testtargs, goodOnes)
        if not good:
            ff.visualize(testimgs, testtargs, badOnes)
    return np.average(np.asarray(errors)), acc
```

```python
def trainEpoch(allbatches, alltargs, imgte, tarte, lrate, dropout, testInt=2, checkInt=10, filename='test-cnn-mnist.pckl'):
    #takes pre-shuffled and separated batches, trains for one epoch
    trloss, traccs, testloss, testaccs = [], [], [], []
    weights = loadweights(filename=filename)
    
    #loops through all batches in epoch
    for b in range(len(allbatches)):
        pkg = train(allbatches[b], alltargs[b], weights, lrate, dropout)   #TODO change back to weights, loss, acc
        
        trloss.append(pkg[1])
        try:
            traccs.append(pkg[2])
        except IndexError:
            print("Got NaN")
            return pkg[0], trloss, traccs, testloss, testaccs, True

        print("[Batch: ", b, " Loss: ", pkg[1], " Accuracy: ", pkg[2], "%]")
        
        if (b % checkInt) == 0:
            with open(filename, 'wb') as f:
                print("Checkpoint, pickling network . . .")
                pickle.dump(pkg[0], f)
        
        if (b % testInt) == 0:
            print("Testing . . .")
            tl, ta = testNet(imgte, tarte, pkg[0], lrate, dropout)
            testloss.append(tl)
            testaccs.append(ta)
            
    return pkg[0], trloss, traccs, testloss, testaccs 
```

```python
def maincnn(trfile, tefile, params, savepaths=['test-cnn-mnist.pckl', 'trlcnn.pckl', 
                                               'telcnn.pckl', 'tracnn.pckl', 'teacnn.pckl']):
    plt.close('all')
    allLoss, testLoss, trAcc, teAcc = [], [], [], []
    
    print("Loading trainng and test data")
    imgtr, tartr, imgte, tarte = ff.make_dataset(trfile, tefile)
    
    #train in epochs
    for i in range(params[0]):    #params[0] = numEpochs
        print("Starting new epoch: ", i)
        #Shuffle training data
        ind, pat, tar = sto.makeBatches(imgtr, tartr, params[1])    #params[1] = batch size
        
        #run stochastic gradient descent on dataset
        #params 2:5 = lrate, dropout, testInt, checkInt
        weights, lo, ac, telo, teac = trainEpoch(pat, tar, imgte, tarte, params[2], params[3], 
                                                 testInt=params[4], checkInt=params[5], filename=savepaths[0])
        allLoss.append(lo)
        testLoss.append(telo)
        trAcc.append(ac)
        try: 
            stahhhp = teac[1]
        except IndexError:
            break
        teAcc.append(teac)
        
    sto.plotsave(allLoss, testLoss, trAcc, teAcc, weights, savepaths)
    minTrLoss = np.amin(np.asarray(allLoss))
    minTeLoss = np.amin(np.asarray(testLoss))
    maxTrAcc = np.amax(np.asarray(trAcc))
    maxTeAcc = np.amax(np.asarray(teAcc))
    return minTrLoss, minTeLoss, maxTrAcc, maxTeAcc
```
